<b style="color:pink; font-size: 1.5em;">Positive Matrix Percentage vs Samples Count</b>

```python
def percent_plotter(max_size, method=2):
    samp_count = np.arange(10, max_size + 1, 10)
    x_list = []
    y_list = []

    for num in samp_count:
        mx = np.linspace(-1, 1, num)
        my = np.linspace(-1, 1, num)
        mz = np.linspace(-1, 1, num)
        y_list.append(percent(mx, my, mz, method))
        x_list.append(num)

    plotter(x_list, y_list, figsize=(6, 4))
```

```python
percent_plotter(400, 3)
```
